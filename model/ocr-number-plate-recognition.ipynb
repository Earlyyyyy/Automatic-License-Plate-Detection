# OCR Number Plate Recognition

Pytesseract is a Python wrapper for the Tesseract OCR (Optical Character Recognition) engine. Tesseract is an open-source OCR engine that was developed by HP and later supported by Google. Pytesseract makes it easier to work with Tesseract from a Python environment.

With Pytesseract, you can easily extract text from an image and convert it into a string that can be processed in a variety of ways. For example, you can use Pytesseract to extract text from scanned documents. This can be useful for a range of applications, such as document digitization, information extraction, and text-based image analysis.

To use Pytesseract, you will first need to install the Tesseract OCR engine. Then, you can install the Pytesseract package using pip. Once you have both installed, you can start using Pytesseract to extract text from images in your Python code.

```python
import os
import cv2
import matplotlib.pyplot as plt
import pytesseract
```

```python
paths=[]
for dirname, _, filenames in os.walk('/kaggle/input/indian-license-plates-with-labels/images'):
    for filename in filenames:
        paths+=[(os.path.join(dirname, filename))]
```

# Some OCR results

```python
for i in [0,1,2,5,16,21,30,36,38]:
    path=paths[i]
    img = cv2.imread(path)
    print(i)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()
    # Perform OCR on the image
    text = pytesseract.image_to_string(img)
    print(text)
```
